### Importing necessary packages

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import pandas as pd
import matplotlib.cm as cm
import scipy.stats as stats

mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'

AxesLabel, TickLabel = [10, 10]
plt.rcParams.update({
    'font.size' : TickLabel,                   # Set font size to 11pt
    'axes.labelsize': AxesLabel,               # -> axis labels
    'legend.fontsize': AxesLabel,              # -> legends
    'xtick.labelsize': TickLabel,              # -> x-axis tick labels
    'ytick.labelsize': TickLabel,              # -> y-axis tick labels
})
```

### Function to find index of the nearest value

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

### Reading .csv file $(f$ file)

```python
df = pd.read_csv(r"turbulence1.csv")

temp = df.to_numpy()
t = temp[:,0]
f = temp[:,1]
g = temp[:,2]
h = temp[:,3]
```

### PDF of $f$ for $0<t<100$

```python
hist, bin_edges = np.histogram(f[0:find_nearest(t, value=100)], bins=200, density=True)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
pdf = hist / np.sum(hist)

plt.bar(bin_centers, hist, width=bin_edges[1] - bin_edges[0], label='PDF of Velocity')
plt.savefig('Plots/PDF1_0t100.pdf', bbox_inches='tight')
```

### PDF of $f$ for $30<t<100$

```python
hist, bin_edges = np.histogram(f[find_nearest(t, value=30):find_nearest(t, value=100)], bins=200, density=True)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
pdf = hist / np.sum(hist)

plt.bar(bin_centers, hist, width=bin_edges[1] - bin_edges[0], label='PDF of Velocity')
plt.savefig('Plots/PDF1_30t100.pdf', bbox_inches='tight')
```

### PDF of $f$ for $30<t<500$

```python
hist, bin_edges = np.histogram(f[find_nearest(t, value=30):find_nearest(t, value=500)], bins=200, density=True)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
pdf = hist / np.sum(hist)

plt.bar(bin_centers, hist, width=bin_edges[1] - bin_edges[0], label='PDF of Velocity')
plt.savefig('Plots/PDF1_30t500.pdf', bbox_inches='tight')
```
